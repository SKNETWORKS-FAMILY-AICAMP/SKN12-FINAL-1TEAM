```python
reader = easyocr.Reader(['ko', 'en'])                             # 추출기 생성(객체)
def extract_text_from_pdf(pdf_path):
  doc = fitz.open(pdf_path)                                       # pdf열기
  full_text = ""                                                  # 텍스트 받을 문자열 변수
  for page_num in range(len(doc)):
    page = doc.load_page(page_num)                                # 여기부터 페이지를 이미지로 변환하는과정
    pix = page.get_pixmap(dpi=300)                                # 페이지별로 구분을 위해 load_page사용, dpi는 해상도(300적당)
    img = Image.open(io.BytesIO(pix.tobytes("png")))              # 변환된이미지를 png로 열기
    result = reader.readtext(np.array(img), detail=0)             # 아까 생성한 추출기로 텍스트추출, 텍스트만 리스트로 추출(detail=0)
    page_text = '\n'.join(result)                                 # 문장 리스트 → 하나의 문자열 변환(줄띄기 단위로) (\n)
    full_text += f"\n\n--- Page {page_num+1} ---\n{page_text}"    # 두줄 띄는 단위로 페이지텍스트들을 병합 (page_num 숫자)가 페이지 타이틀로

  return full_text
```

```python
full_text = extract_text_from_pdf(pdf_path)
```
